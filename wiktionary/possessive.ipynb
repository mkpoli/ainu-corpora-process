```python
import csv
from utils.text import remove_accent

with open('../../dictionary/input/possessive.tsv', 'r') as f:
    possessives = [row[0:3] for row in csv.reader(f, delimiter='\t')]
possessives.pop(0)
# remove accent
possessives = [[remove_accent(word) for word in row] for row in possessives]
possessives
```

```python
from collections import defaultdict
duplicated_possessives = defaultdict(set)
for i, (orig, short, long, gloss) in enumerate(possessives):
    for j, (orig2, short2, long2, gloss2) in enumerate(possessives):
        if i != j and orig == orig2:
            duplicated_possessives[orig].add((short2, long2, gloss2))
duplicated_possessives
```

```python
from collections import defaultdict
from Language.linguistics.ainu.wiktionary.template import create_multiple_pos_word, Lemma

#     sa -> 姉　 -> 0. (sa, saha)
#        -> 節　 -> 0. (sake, sakehe)
# corpok -> 下　 -> 0. (corpokke, corpokkehe)
#           　　 -> 1. (corpokike, corpokikehe)
#  utur -> 間　 -> 0. (uturu, uturuke)
#          　　 -> 1. (uturke, uturkehe)
#       -> 下座 -> 0. (uturu, uturuke)
#          　　 -> 1. (uturke, uturkehe)

lemma_map: dict[str, dict[str, set[str]]] = defaultdict(lambda: defaultdict(set))

for word, short, long, gloss in possessives:
    lemma_map[word][gloss].add(short)
    lemma_map[word][gloss].add(long)

lemma_contents: dict[str, str] = {}

for lemma, glosses in lemma_map.items():
    
    lemma_contents[lemma] = create_multiple_pos_word(
        [
            Lemma(
                lemma,
                "noun",
                [add_furigana(gloss)],
                [],
                {
                    "所属形": sorted(list(forms), key=lambda x: len(x)),
                },
                []
            ) for gloss, forms in glosses.items()
        ]
    )
    
with open('_lemma_contents.txt', 'w') as f:
    for lemma, content in lemma_contents.items():
        f.write("\n")
        f.write("\n")
        f.write(f"== {lemma} ==\n")
        f.write(content)
        f.write("\n")
        f.write("\n")

```

```python
from Language.linguistics.ainu.wiktionary.template import noun

from dataclasses import dataclass
@dataclass
class Entry:
    lemmas: list[str]
    content: str

entries: dict[str, Entry] = {}

for form, pairs in form_map.items():
    # possese
    
    form_of = [
        f'{{{{noun form of|ain|{word}||所属形|tr={{{{ain-kana-conv|{word}}}}}|t={gloss}}}}}'
        for word, gloss in pairs
    ]
    
    lemmas = list(set(word for word, _ in pairs))
    
    result = noun(form_of, has_ref=False, cats=['名詞 所属形'], no_period=True)
    entries[form] = Entry(lemmas=lemmas, content=result)

for form, entry in sorted(entries.items(), key=lambda x: x[0]):
    print(f'{form:>20} : {entry.lemmas}')
```
