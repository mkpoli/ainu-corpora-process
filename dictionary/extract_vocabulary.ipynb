# 各辞書の単語一覧を取得

## 語彙集

### 語彙集データを取得

```python
# Itak ueroskip
import requests

req = requests.get("https://itak.aynu.org/api/gdoc")

itakuoeroskip = req.json()["table"]

print(itakuoeroskip)
```

### 語彙集データからアイヌ語単語を抽出

```python
import regex as re
import datetime

words = set()
for i in itakuoeroskip:
    for expression in i.get("Aynu", "").split(", "):
        for w in expression.split(" "):
            if re.match(r"^(V|N|S)\d$", w):
                continue
            w = re.sub(r"(V|N|S)\d", "", w)
            w = re.sub(r"(HH|MM|SS|YYYY|DD)", "", w)
            w = re.sub(r"(\d+)", "", w)
            w = re.sub(r"[!?.,\";]+", "", w)
            w = w.lower()

            if w.startswith("[") or w.endswith("]"):
                continue
            if w.startswith("(") or w.endswith(")"):
                continue
            if not w:
                continue
            if re.match(r"^[\p{P}\s+~]+$", w):
                continue

            if re.match(r"^[\p{scx=Kana}\p{scx=Hiragana}]+$", w):
                continue
            
            if "{" and "}" in w:
                # kes{pa,to} > [kespa, kesto]
                for ww in w.split("{")[1].split("}")[0].split(","):
                    words.add(w.split("{")[0] + ww)
            elif "=" in w:
                if w.startswith("a="):
                    words.add(w.split("=")[1])
                elif w.endswith("=an"):
                    words.add(w[:-2])
            else:
                words.add(w)

words = set(w.replace("=", "") for w in words)
print(len(words))

for w in words:
    if re.match(r"[^a-zA-Z]", w):
        print(w)


current_time_utc_9 = datetime.datetime.now(datetime.timezone(datetime.timedelta(hours=9)))
with open(f"output/itakuoeroskip_ainu_words_{current_time_utc_9.strftime('%Y%m%d%H%M%S')}.txt", "w") as f:
    for w in sorted(words):
        f.write(w + "\n")
```
