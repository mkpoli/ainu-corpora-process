# Generate cleaned word frequency for other use

## Load data

```python
import csv
with open("output/ainu_words_hokkaido_traditional.tsv", "r", encoding="utf-8") as f:
    csv_reader = csv.reader(f, delimiter="\t")
    lines = [(row[0], int(row[1])) for row in csv_reader]

print(len(lines))
for line in lines[:20]:
    print(line)
```

## Clean data

```python
from utils.tokenize import is_word
from utils.lemmatize import normalize

valid_words: list[tuple[str, int]] = []
for line in lines:
    if is_word(line[0]):
        lemma = normalize(line[0])
        if lemma:
            valid_words.append((lemma, line[1]))

print(len(valid_words))
for word in valid_words[:20]:
    print(word)
```

## Export data

```python
import csv
from pathlib import Path


OUTPUT_DIR = Path("output/frequency")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

with open(OUTPUT_DIR / "ainu_words_hokkaido_traditional_cleaned.tsv", "w", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerow(["word", "freq"])
    writer.writerows(valid_words)

with open(OUTPUT_DIR / "ainu_words_hokkaido_traditional_cleaned_2_plus.tsv", "w", encoding="utf-8") as f:
    writer = csv.writer(f, delimiter="\t")
    writer.writerow(["word", "freq"])
    for word, freq in valid_words:
        if freq >= 2:
            writer.writerow([word, freq])
```
