```python
import json
from pathlib import Path



def create_conllu_entry(sentence):
    # Extract necessary fields from the JSON data
    text = sentence.get("text", "")
    gloss_jp = sentence.get("gloss_jp", "")
    translation_jp = sentence.get("translation_jp", "")

    # Create CoNLL-U format lines
    lines = [
        f"# text = {' '.join(text.split('||'))}",
        f"# translation_jp = {translation_jp}",
    ]

    # Split the text and gloss into tokens
    text_tokens = text.split("||")
    gloss_tokens = gloss_jp.split("||")

    # Create CoNLL-U entries for each token
    for idx, (token, gloss) in enumerate(zip(text_tokens, gloss_tokens), start=1):
        lines.append(f"{idx}\t{token}\t_\t_\t_\tGloss={gloss}\t_\t_\t_\t_")

    return "\n".join(lines) + "\n"


# Process each JSON file in the directory
for json_file in json_dir.glob("*.json"):
    with open(json_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Create a filename based on the JSON file name
    output_filename = json_file.stem + ".conllu"

    # Open the output file for writing
    with open(output_dir / output_filename, "w", encoding="utf-8") as f:
        # Process each sentence in the JSON data
        for sentence in data["data"]:
            conllu_content = create_conllu_entry(sentence)
            f.write(conllu_content)
            f.write("\n")  # Add a blank line between sentences
```
