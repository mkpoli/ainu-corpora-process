```python
import requests
import json

API_URL = "https://ainu.ninjal.ac.jp/folklore/jp/stories/?_dc=1735735687703&page=1&start=0&limit=25"
API_URL_2 = "https://ainu.ninjal.ac.jp/folklore/jp/stories_download/?_dc=1735735687703&page=1&start=0&limit=25"

with requests.get(API_URL) as response:
    response.raise_for_status()
    data = json.loads(response.text)

with requests.get(API_URL_2) as response:
    response.raise_for_status()
    data_2 = json.loads(response.text)

print(data)
print(data_2)
```

```python
len(data["data"])
```

```python
story_ids = [story["story_id"] for story in data["data"]]

story_ids


```

```python
urls = [
    f"https://ainu.ninjal.ac.jp/folklore/jp/story_body/{story_id}/"
    # ?_dc=1735735687703&page=1&start=0&limit=25
    for story_id in story_ids
]
for url in urls[:10]:
    print(url)
```

```python

with requests.get(urls[0]) as response:
    response.raise_for_status()
    data = json.loads(response.text)
from time import sleep
from tqdm.notebook import tqdm

from pathlib import Path

OUTPUT_DIR = Path("../output/ninjal/")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


for id in tqdm(story_ids):
    if (OUTPUT_DIR / f"{id}.json").exists():
        print(f"Skipping {id} because it already exists")
        continue
    with requests.get(f"https://ainu.ninjal.ac.jp/folklore/jp/story_body/{id}/") as response:
        response.raise_for_status()
        data = json.loads(response.text)
    with open(OUTPUT_DIR / f"{id}.json", "w") as f:
        json.dump(data, f, ensure_ascii=False)
    sleep(1)
```
