## Inspect words

```python
import csv
import regex as re

with open(
    "./output/ainu_words_hokkaido_traditional.tsv", "r", encoding="utf-8"
) as f:
    csv_reader = csv.reader(f, delimiter="\t")
    lines = [row for row in csv_reader]

with_long_vowel = set()
for line in lines:
    if re.match(r"[âîûêô]", line[0]):
        print(line[0])
        with_long_vowel.add(line[0])
print(len(with_long_vowel))
for word in sorted(with_long_vowel):
    print(word)
```
